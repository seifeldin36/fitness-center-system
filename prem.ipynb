```python
import pandas as pd
```

```python
data = pd.read_csv('/content/player_stats.csv')
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
mask = data['nation'].isna()
data[mask]
```

```python
# Select rows where the name is 'Jake Evans' and fill the missing 'nation' values
data.loc[data['name'] == 'Jake Evans', 'nation'] = data.loc[data['name'] == 'Jake Evans', 'nation'].fillna('England')
```

```python
data.loc[data['name']== 'Mateus Mane','nation'] = data.loc[data['name']== 'Mateus Mane','nation'].fillna('Portugal')
```

```python
data.loc[data['name']=='Jeremy Monga','nation'] = data.loc[data['name']=='Jeremy Monga','nation'].fillna('England')
```

```python
data.info()
```

```python
mask_age = data['age'].isna()
data[mask_age]
```

```python
# Select rows where the name is 'Jake Evans' and fill the missing 'nation' values
data.loc[data['name'] == 'Jake Evans', 'age'] = data.loc[data['name'] == 'Jake Evans', 'age'].fillna('17')
data.loc[data['name']== 'Mateus Mane','age'] = data.loc[data['name']== 'Mateus Mane','age'].fillna('17')
data.loc[data['name']=='Jeremy Monga','age'] = data.loc[data['name']=='Jeremy Monga','age'].fillna('16')
```

```python
data.info()
```

```python
# Select rows where the name is 'Jake Evans' and fill the missing 'nation' values
data.loc[data['name'] == 'Jake Evans', 'born'] = data.loc[data['name'] == 'Jake Evans', 'born'].fillna('2008')
data.loc[data['name']== 'Mateus Mane','born'] = data.loc[data['name']== 'Mateus Mane','born'].fillna('2008')
data.loc[data['name']=='Jeremy Monga','born'] = data.loc[data['name']=='Jeremy Monga','born'].fillna('2009')
```

```python
data.info()
```

```python
data.head()
```

```python
#Show how many goals for each team "descinding oreder"
data.groupby('team')['goals'].sum().sort_values(ascending=False)
```

```python
#Show The top 10 player scored goals
data.groupby('name')['goals'].sum().sort_values(ascending=False).head(10)
```

```python
#Show The top 10 players made assists
data.groupby('name')['assists'].sum().sort_values(ascending=False).head(10)
```

```python
#Show The top 10 players got red card
data.groupby('name')['red'].sum().sort_values(ascending=False).head(10)
```

```python
#Show The top 10 scored by penalty
data.groupby('name')['penalty_kicks'].sum().sort_values(ascending=False).head(10)
```

```python
#Show The top 10 players got red card
data.groupby('name')[['penalty_kick_attempts','penalty_kicks','team']].sum().sort_values(by= 'penalty_kick_attempts',ascending=False).head(10)
```

```python
data.info()
```

```python
data.groupby('name')[['minutes','position','team']].sum().sort_values(by= 'minutes',ascending=False).head(10)
```
